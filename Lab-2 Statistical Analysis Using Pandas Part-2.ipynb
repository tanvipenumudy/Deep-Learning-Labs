```python
import numpy as np
import pandas as pd
```

```python
#1
df=pd.read_csv("https://github.com/JWarmenhoven/ISLR-python/raw/master/Notebooks/Data/Advertising.csv")
df
```

```python
#2
df.isnull()
```

```python
#3
df1=pd.DataFrame()
df1['Sales Type']=np.select([df['Sales']>=15,(df['Sales']<15)&(df['Sales']>=5),df['Sales']<5],['high','medium','low'])
df1
```

```python
#4
data=pd.concat([df,df1],sort=False,axis=1)
data
```

```python
#5
print('Sum:\n',data.groupby('Sales Type')['TV'].sum())
print('\nMean:\n',data.groupby('Sales Type')['TV'].mean())
print('\nSum:\n',data.groupby('Sales Type')['Newspaper'].sum())
print('\nMean:\n',data.groupby('Sales Type')['Newspaper'].mean())
print('\nSum:\n',data.groupby('Sales Type')['Radio'].sum())
print('\nMean:\n',data.groupby('Sales Type')['Radio'].mean())
```

```python
#6
df1=data.pivot(index='Unnamed: 0',columns='Sales Type',values='Sales')
df1
```

```python
#7
#Replacing every 6th row of TV and 10th row of Newspaper with NaN
dfc=df.copy()
for i in range(len(dfc)):
    if((i+1)%6==0):
        dfc['TV'][i]=np.nan
    if((i+1)%10==0):
        dfc['Newspaper'][i]=np.nan
```

```python
dfc
```

```python
#printing df rows that have no NaN values
dfn=dfc.dropna()
dfn
```

```python
#printing rows with one or more NaN values
dfna=dfc[dfc.isna().any(axis=1)]
dfna
```

```python
#8
#Filling missing values with 0
df2=dfc.fillna(0)
df2
```

```python
#Filling missing values with avg of column
df3=dfc.apply(lambda x: x.fillna(x.mean()),axis=0)
df3
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df4=pd.DataFrame(columns=['TV','Radio','Newspaper'])
df4['TV']=df['TV']
df4['Radio']=df['Radio']
df4['Newspaper']=df['Newspaper']
df4
```

```python
from sklearn import preprocessing
```

```python
#9
x=df4.values
mm=preprocessing.MinMaxScaler()
x1=mm.fit_transform(x)
df4=pd.DataFrame(x1)
df4
```

```python
df['TV']=df4[0]
df['Radio']=df4[1]
df['Newspaper']=df4[2]
df
```

```python
#10
from sklearn.model_selection import train_test_split
X=df.drop('Sales',axis=1).values
y=df['Sales'].values
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.6,test_size=0.4,random_state=123)
```

```python
#11
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train,y_train)
```

```python
#12
predictions=model.predict(X_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions)
predictions=model.predict(X_test)
plt.xlabel('True values')
plt.ylabel('Predictions')
plt.show()
```

```python
#MSE
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions)
print(f"MSE: {mse}")
```

```python
#R2 Score
from sklearn.metrics import r2_score
r2=r2_score(y_test,predictions)
print(f"R2: {r2}")
```

```python
from sklearn.model_selection import learning_curve
train_sizes=[5,10,20,40,80,160]
train_sizes,train_scores,validation_scores=learning_curve(
estimator=LinearRegression(),
X=df.drop('Sales',axis=1).values,
y=df['Sales'].values, train_sizes = train_sizes,cv=5,
scoring='neg_mean_squared_error')

```

```python
print('Training scores:\n\n',train_scores)
print('\n','-'*70) # separator to make the output easy to read
print('\nValidation scores:\n\n',validation_scores)
```

```python
train_scores_mean=-train_scores.mean(axis=1)
validation_scores_mean=-validation_scores.mean(axis=1)
print('Mean training scores\n\n',pd.Series(train_scores_mean,index=train_sizes))
print('\n','-'*20) #separator
print('\nMean validation scores\n\n',pd.Series(validation_scores_mean,index=train_sizes))
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.plot(train_sizes,train_scores_mean,label='Training error')
plt.plot(train_sizes,validation_scores_mean,label='Validation error')
plt.ylabel('MSE',fontsize=14)
plt.xlabel('Training set size',fontsize=14)
plt.title('Learning curves for a linear regression model',fontsize=18,y=1.03)
plt.legend()
plt.ylim(0,10)
```
